```python
import sklearn
from sklearn import datasets
from sklearn import svm
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score   
```

```python
cancer_data = datasets.load_breast_cancer()
data = pd.DataFrame(data = cancer_data.data, columns= cancer_data.feature_names)
data.head()
```

```python
X = cancer_data.data
y = cancer_data.target
X_train,X_test,y_train,y_test = train_test_split(X,y, train_size=0.7, random_state=42)
classes = ['malignant', 'benign ']
```

```python
svm_model = svm.SVC(kernel="linear")
svm_model.fit(X_train, y_train)
pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, pred) * 100

print("\nAccuracy of the model is: " + str(round(accuracy, 2)) + "%\n")
```

```python
count = 0
for i in range(len(pred)):
    if classes[pred[i]] == classes[y_test[i]]:
        count+=1
    print("Predicted: ", classes[pred[i]], "\nData: ", X_test[i], "\nActual: ", classes[y_test[i]], "\n")
```
