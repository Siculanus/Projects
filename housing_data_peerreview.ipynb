## Loading the Dataset

```python
house_df = pd.read_csv('SampleDatasetHousing_Data - housing_data.csv')
house_df
```

```python
house_df.sort_values('price')
```

```python
house_df = house_df.to_numpy()
house_df
```

## Converting the data to matrices for calculation; separating target from predictor

```python
ones = np.ones((100,1))
ones = Matrix(ones)
```

```python
X = Matrix(house_df[:,0])
X = X.col_insert(0, ones)
X
```

```python
Y = Matrix(house_df[:,1])
Y
```

## Applying the normal equation

```python
P1 = X.T*X
P1
```

```python
P = P1.inv()
P
```

```python
Q = X.T*Y
Q
```

```python
A = P*Q
A
```
